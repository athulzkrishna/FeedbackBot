```python
# Using opencv to access the webcam
import docx
mydoc = docx.Document()
import cv2 
def capturing(name,content,rating):
    say='D:\images'+'\\'+str(name)+'.jpg'
    key = cv2. waitKey(1)
    webcam = cv2.VideoCapture(0)
    while True:
        try:
            check, frame = webcam.read()
            print(check) #prints true as long as the webcam is running
            print(frame) #prints matrix values of each framecd 
            cv2.imshow("Capturing", frame)
            key = cv2.waitKey(1)
            if key == ord('s'): 
                cv2.imwrite(filename=say, img=frame)
                webcam.release()
                img_new = cv2.imread(say, cv2.IMREAD_GRAYSCALE)
                img_new = cv2.imshow("Captured Image", img_new)
                cv2.waitKey(1650)
                cv2.destroyAllWindows()
                
                print("Image saved!")

                break
            elif key == ord('q'):
                print("Turning off camera.")
                webcam.release()
                print("Camera off.")
                print("Program ended.")
                cv2.destroyAllWindows()
                break

        except(KeyboardInterrupt):
            print("Turning off camera.")
            webcam.release()
            print("Camera off.")
            print("Program ended.")
            cv2.destroyAllWindows()
            break
    mydoc = docx.Document("D:\my_written_file.docx")
    mydoc.add_heading(name, 0)
    mydoc.add_heading('FEEDBACK:', 1)
    mydoc.add_paragraph(content)
    mydoc.add_heading('RATING', 1)
    mydoc.add_paragraph(rating)
    mydoc.add_picture(say, width=docx.shared.Inches(5), height=docx.shared.Inches(5))
    mydoc.save("D:\my_written_file.docx") 
```

```python
#summary making function
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
def make_summary(text):
    stopwords=list(STOP_WORDS)
    #stopwords
    nlp=spacy.load('en_core_web_sm')
    doc=nlp(text)
    tokens=[token.text for token in doc]
    global punctuation
    punctuation=punctuation+'\n'
    freq={}
    for w in doc:
        if w.text.lower() not in stopwords:
            if w.text.lower() not in punctuation:
                if w.text.lower() not in freq.keys():
                    freq[w.text.lower()]=1
                else:
                    freq[w.text.lower()]+=1
    maxi=max(freq.values())
    for i in freq.keys():
        freq[i]=freq[i]/maxi
    sentence_tokens=[sent for sent in doc.sents]
    scores={}
    for sent in sentence_tokens:
        for word in sent:
            if word.text.lower() in freq.keys():
                if sent not in scores.keys():
                    scores[sent]=freq[word.text.lower()]
                else:
                    scores[sent]+=freq[word.text.lower()]
    from heapq import nlargest
    length=int(len(sentence_tokens)*0.4)
    summary=nlargest(length,scores,key=scores.get)
    final_summary=[word.text for word in summary]
    summary=''.join(final_summary)
    #print(len(summary))
    return summary
```
